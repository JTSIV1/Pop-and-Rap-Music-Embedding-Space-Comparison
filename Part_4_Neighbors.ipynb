### Neighbors for Pop and Rap

Imports

```python
import gensim.downloader as api
from gensim.models import Word2Vec
from random import choices
```

Load Models

```python
model_pop = Word2Vec.load('pop_sg.model')
model_rap = Word2Vec.load('rap_sg.model')
```

Get Vocab and Choose Random Words

```python
# Get all words using key_to_index
all_words_pop = list(model_pop.wv.key_to_index.keys())
all_words_rap = list(model_rap.wv.key_to_index.keys())

# Find the intersection of words using set operations
all_words = list(set(all_words_pop) & set(all_words_rap))
random_words = choices(all_words, k=20)
```

Nearest Neighbors

```python
def jaccard_similarity(list1, list2):
  intersection = len(set(list1).intersection(set(list2)))
  union = len(set(list1).union(set(list2)))
  if union == 0:
    return 0
  return float(intersection) / union
```

```python
sum = 0.0
for word in random_words:
  similar_words = model_pop.wv.most_similar(word, topn=20)
  words1 = []
  print(f"(Pop) Nearest neighbors for '{word}':")
  for neighbor, similarity in similar_words:
    print(f"\t- '{neighbor}': {similarity:.4f}")
    words1.append(neighbor)
  similar_words = model_rap.wv.most_similar(word, topn=20)
  words2 = []
  print(f"(Rap) Nearest neighbors for '{word}':")
  for neighbor, similarity in similar_words:
    print(f"\t- '{neighbor}': {similarity:.4f}")
    words2.append(neighbor)
  print(f"Jaccard Similarity for {word}: {jaccard_similarity(words1, words2)}")
  sum = sum + jaccard_similarity(words1, words2)

print(f"Average Jaccard Similarity: {sum / len(random_words)}")
```
